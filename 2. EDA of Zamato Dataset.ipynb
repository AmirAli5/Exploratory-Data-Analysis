```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.heatmap(df.isnull(), yticklabels= False, cbar= False, cmap = 'viridis')
```

```python
df_country = pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
final_df.head()
```

```python
### To check Datatype
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_value = final_df.Country.value_counts().values
country_value
```

```python
# Pie_chart top three country
plt.pie(country_value[:3], labels = country_names[:3], autopct = "%1.2f%%")
plt.show()
```

Observation: Zamato Maxmium records or transaction are from India After that USA and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---------> excellent
2. When rating is between 4.0 to 4.4 ---------> very good
4. When rating is between 3.5.0 to 3.9 ---------> good
3. When rating is between 2.5 to 3.4 ---------> average
5. When rating is between 1.8 to 2.4 ---------> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count',  data = ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color',  data = ratings, palette = ['black', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observation: <br>
    1. Not Rated count is very high <br>
    2. Maximum number of rating are between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x = "Rating color", data = ratings,  palette = ['black', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
### Find the countires name that has given 0 ratings
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

Observations <br>
Maxmium number of 0 ratings are from Indian customers

```python
### find out which currency is used by which country
final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
## Which country do have online deliveries option
final_df[final_df['Has Online delivery'] == 'Yes'].Country.value_counts()
```

Observation
1. Online Deveries are available in india and UAE
